```python
import numpy as np
```

```python
values = np.array([1, 2, 3, 4, 5, 6])
```

```python
# Iteration
# O(n) time
# O(1) space
def dynamic_dice_game(n):
    ev = values.mean()
    i = 1 # strat from round 1
    while i < n:
        ev = np.mean(np.maximum(ev, values))
        i += 1
    return ev

# Recursion
# O(n) time
# O(n) space
def recursive_dynamic_dice_game(n):
    if n == 1:
        return values.mean()
    ev = recursive_dynamic_dice_game(n - 1)
    return np.mean(np.maximum(values, ev))
    # return np.sum([1/6 * max(v, ev) for v in values])
    
    # mask = values > ev
    # greater_values = values[mask]
    # return 1/6 * np.sum(greater_values)  + (1 - len(greater_values) / 6) * ev

# test:
print(dynamic_dice_game(3)==14/3)
print(recursive_dynamic_dice_game(3)==14/3)
```
